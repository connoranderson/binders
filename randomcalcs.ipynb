```python
import numpy as np
import matplotlib.pyplot  as plt
```

```python
# Customer inputs
valet_hourly_wage = 19.5
num_shifts = 2
customer_payback_period_months = 36

# Inputs

payback_period_months = 24

robot_cost = 35000
num_robots = 3
bay_cost = 35000 + 1200
num_bays = 1
deployment_cost = 10000
tray_nominal_cost = 3500
num_trays = 39


```

```python
customer_op_ex_savings = np.array(np.linspace(0,1,100))

cust_cost_monthly = (1-customer_op_ex_savings) * valet_hourly_wage * 8 * num_shifts * 30
cust_capx = customer_payback_period_months * customer_op_ex_savings * valet_hourly_wage * 8 * num_shifts * 30  # Assume 2 yr payback

volley_non_tray_costs = num_robots * robot_cost + num_bays * bay_cost + deployment_cost

tray_cost = (payback_period_months * cust_cost_monthly + cust_capx - volley_non_tray_costs) / num_trays
```

```python
plt.plot(customer_op_ex_savings, tray_cost)
```

```python
customer_op_ex_savings = np.array(np.linspace(0,1,100))

cust_cost_monthly = (1-customer_op_ex_savings) * valet_hourly_wage * 8 * num_shifts * 30
cust_capx = customer_payback_period_months * customer_op_ex_savings * valet_hourly_wage * 8 * num_shifts * 30  # Assume 2 yr payback

costs = num_trays*tray_nominal_cost + num_robots * robot_cost + num_bays * bay_cost + deployment_cost

payback_period = (costs - cust_capx)/cust_cost_monthly
```

```python
plt.plot(customer_op_ex_savings, payback_period)
plt.ylim(0,80)
```
